```python
from langchain.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
import transformers
import torch
import warnings
warnings.filterwarnings('ignore')
```

```python
#Model to use : meta-llama/Llama-2-7b-chat-hf
#Need access to it so fill up the form on huggingface first
```

```python
model="meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
#text generation model
#torch_dtype : precision of the model i.e. how to store weights
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt="What would be a good name for a company that makes Electric Cars ?"
```

```python
print(llm(prompt))
```

```python
prompt="I want to open a restaurant for  Japanese food. Suggest me a traditional name for it"
```

```python
print(llm(prompt))
```

```python
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
#example 1
```

```python
prompt_template1=PromptTemplate(input_variables=["cuisine"],
                               template="I want to open a restaurant for {cuisine} food. Suggest a traditional name for this")
```

```python
input_prompt1=prompt_template1.format(cuisine="Punjabi")
```

```python
print(input_prompt1)
```

```python
#example 2
```

```python
prompt_template2=PromptTemplate(input_variables=["book_name"],
                               template="Provide me a concise summary in 100 words of the book {book_name}")
```

```python
input_prompt2=prompt_template2.format(book_name="And Then There Were None")
```

```python
print(input_prompt2)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template1, verbose=True)
response= chain.run("Mughlai")
print(response)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template2, verbose=True)
response= chain.run("Be Here Now")
print(response)
```
